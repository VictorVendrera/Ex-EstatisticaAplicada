```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
#--------------------------------------Exercicios Distribuição de frequência----------------------------------------------------
#Exericico 1
acidentesDiarios50d = [6,9,2,7,0,8,2,5,4,2,5,4,4,4,4,2,5,6,3,7,3,8,8,4,4,4,7,7,6,5,4,7,5,3,3,1,3,8,0,6,5,1,2,3,3,0,5,6,6,3]
```

```python
#Tranformando o vetor em dicionário distribuido
numeroD = Counter(acidentesDiarios50d)
numeroD
```

```python
plt.bar(numeroD.keys(), numeroD.values())
plt.ylabel("Número de Acidentes")
plt.xlabel("Dias")
plt.title("Exercicio1-Distribuição de frequência")
plt.show()
```
