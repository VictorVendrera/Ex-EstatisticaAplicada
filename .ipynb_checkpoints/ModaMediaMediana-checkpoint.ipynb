```python
import pandas as pd
import numpy as np
import statistics
from collections import Counter
```

```python
#-----------------------------------------------------Exercicio1----------------------------------------------------------------
pesos= [3,4,3.5,5,3.5,4,5,5.5,4,5]
media= statistics.mean(pesos)
moda = statistics.mode(pesos)
mediana = statistics.median(pesos)
print("Média:",media, "Moda:",moda, "Mediana:", mediana)
```

```python

#-----------------------------------------------------Atividade3---------------------------------------------------------------
## Lista e variavel auxiliar
x = []
soma = 0
soma1 = 0
soma2 = 0
#Criando o DataFrame
data = np.array([["0|---50",10],
                 ["50|---100",28],
                 ["100|---150",12],
                 ["150|---200",2],
                 ["200|---250",1],
                 ["250|---300",1]])
df = pd.DataFrame(data,columns=["Consumo por nota","Nª de notas (fi)"])


df["xi"]=[25,75,125,175,225,275] #Metade da primeira coluna resolvi fazer manual pois não vi mta necessidade de codar

df['Nª de notas (fi)']=df['Nª de notas (fi)'].astype("float64")
#Crinando xi.fi
df["xi.fi"] = df['Nª de notas (fi)']* df['xi']

#Criando F
for i in df['Nª de notas (fi)']:    
    soma += i
    x.append(soma)
df['F']=x 

#Somando o total
for i in df['Nª de notas (fi)']:
    soma1 += i 
for i in df['xi.fi']:
    soma2 += i

#Adicionando total
df.loc[6] = ['Total:',soma1,'',soma2,'']
df

```

```python
##Respostas3
df['Nª de notas (fi)']=df['Nª de notas (fi)'].astype("int64")
df['xi.fi']= df['xi.fi'].astype("int64")
media = round(df.iloc[6,3]/df.iloc[6,1],2) #Divisão de 4700/54
moda = round(df.iloc[1,2],2)#Número mais comum
mediana = round(50+(27-10)*50/28,2) #Tive que fazer de modo manual mas a tabela foi de grande ajuda
print("Média:",media, "Moda:",moda, "Mediana:", mediana)
```
